```python
def get_embedding_raw_df():
    return pd.read_pickle(os.path.join('visualizer_data', 'activations', 'painters_ds', 'embedding-raw.pkl'))

def get_embedding_raw_sample(sample_index): # will be sample
    df_raw = pd.read_pickle(os.path.join('visualizer_data', 'activations', 'painters_ds', 'embedding-raw.pkl'))
    df_raw_sample = df_raw[df_raw.index == sample_index].iloc[:, 4:]
    
    arr = df_raw_sample.to_numpy().reshape(-1, 64)
    
    return pd.DataFrame(arr)

def standardize_from_file(df, layer_name):
    stats_path = os.path.join('visualizer_data', 'activations', 'painters_ds', 'stats', f'{layer_name}-describe.pkl')
    layer_stats = pd.read_pickle(stats_path)
    df_s = df.copy()
    for col in df:
        df_s[col] = (df[col] - layer_stats.loc['mean', f"neuron_{col+1}"] / layer_stats.loc['std', f"neuron_{col+1}"])
        
    return df_s

def standardize(df):
    stats_path = os.path.join('visualizer_data', 'activations', 'painters_ds', 'stats')
    df_s = df.copy()
    df_s.describe().to_pickle(os.path.join(stats_path, f"embedding-describe.pkl"))
    for col in df:
        df_s[col] = (df[col] - df[col].mean()) / df[col].std()
    
    return df_s

def get_mean_df_per_neuron(df):
    return pd.DataFrame(pd.DataFrame(df).mean()).T
```

```python
df_sr = get_embedding_raw_sample('http://dbpedia.org/resource/Philip_Pearlstein') # Sample Raw
df_sr
```

```python
layer_name = "embedding"
df_ssr = standardize_from_file(df_sr, layer_name) # Standardisation du Sample Raw
df_ssr
```

```python
df_mssr = get_mean_df_per_neuron(df_ssr) # Moyenne de la Standardisation du Sample Raw
df_mssr
```

Wanting to standardize then mean the samples in embedding

Load the raw embeddings

```python
er_path = os.path.join('visualizer_data', 'activations', 'painters_ds', 'embedding-raw.pkl')
df_fsr = pd.read_pickle(er_path).iloc[:, 4:] # df Full Sample Raw
df_fsr
```

Then standardize the whole df

```python
df_fssr = standardize(df_fsr)
df_fssr
```

For each sample:
* reshape -1, 64 to get a 150x64 matrix
* then apply mean of pandas on this matrix to get a 1x64 matrix

```python
matrix = df_fssr.to_numpy()
output_dim = 64 # will be dynamic instead of static
acts = []
for i in range(len(matrix)):
    acts.append(pd.DataFrame(pd.DataFrame(matrix[i].reshape(-1, output_dim)).mean()).T)
```

```python
activations = np.array(acts).reshape(-1, output_dim) # Because acts was 6576x1x64
activations.shape
```

```python
new_df = pd.DataFrame()
for neuron_index, value_list in enumerate(activations.T):
    index = f"neuron_{neuron_index+1}"
    new_df[index] = value_list
```

```python
new_df.set_index(df_fssr.index, inplace=True)
new_df
```
