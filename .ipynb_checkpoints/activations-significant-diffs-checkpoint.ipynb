```python
import pandas as pd
import numpy as np

df_activations = pd.read_csv('./data/activations.csv')
df_activations.shape
```

```python
import glob as glob
import plotly.express as px
```

```python
df_c = pd.DataFrame() # Initializing DF
cat_activations = [] # Array which will contain a DF for each category

for filename in glob.glob('data/activations_*.csv'):
    print(f"Concatening {filename} to df_c...")
    df_temp = pd.read_csv(filename)
    cat_activations.append(df_temp) # Saving each DF part into an array
    print(f"    => as category : {list(df_temp.columns)[0]}")
    df_c = pd.concat([df_c, pd.read_csv(filename)], axis=1) # Bulding a whole DF
```

## **Need to find a way to also save category's name while saving activations as a csv file**

```python
for activ in cat_activations:
    fig = px.scatter(activ, title=f"Mean activations for {activ.columns[0]}")
    fig.show()
```

```python
def get_activation_for_cat(category):
    for activ in cat_activations:
        if activ.columns[0] == category:
            return activ
        
    return None
```

## Building Heatmaps for each category

```python
for activ in cat_activations:
    fig = px.imshow(activ.T, title=f"Heatmap for {activ.columns[0]}", height=350)
    fig.show()
```

```python
def get_act_heatmap_not_for_cat(category):
    dict_act = {}
    
    df = pd.DataFrame()
    for cat in cat_activations:
        for index, row in cat.iterrows():
            if f"neuron_{index + 1}" not in dict_act.keys():
                dict_act[f"neuron_{index + 1}"] = []
            dict_act[f"neuron_{index + 1}"].append(row[0])
            
    # Plot heatmaps
    arr_mean_act = pd.DataFrame().from_dict(dict_act)
    df_mean_act = pd.DataFrame(arr_mean_act.mean())
    
    fig = px.imshow(get_activation_for_cat(category).T.reset_index().rename(columns={'index': ''}), 
                    title=f"Heatmap for category {category}")
    fig.show()
    
    fig = px.imshow(df_mean_act.T, title=f"Heatmap for all categories except {category}", height=350)
    fig.show()
    
```

```python
get_act_heatmap_not_for_cat('American_black-and-white_films')
```
